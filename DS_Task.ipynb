```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Task 2
# House Price Prediction using Machine Learning
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_excel("HousePricePrediction.xlsx")

print(dataset.head(5))

dataset.shape
```

```
obj = (dataset.dtypes == 'object')
object_cols = list(obj[obj].index)
print("Categorical variables:",len(object_cols))

int_ = (dataset.dtypes == 'int')
num_cols = list(int_[int_].index)
print("Integer variables:",len(num_cols))

fl = (dataset.dtypes == 'float')
fl_cols = list(fl[fl].index)
print("Float variables:",len(fl_cols))
```

```
# Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

numeric_data = dataset.select_dtypes(include=['number'])

corr_matrix = numeric_data.corr()

plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix,
            cmap='BrBG',
            fmt='.2f',
            linewidths=2,
            annot=True)
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

object_cols = [col for col in dataset.columns if dataset[col].dtype == 'object']

unique_values = [dataset[col].nunique() for col in object_cols]

if len(object_cols) == len(unique_values):
    plt.figure(figsize=(12, 6))
    sns.barplot(x=object_cols, y=unique_values)
    plt.xlabel('Categorical Features')
    plt.ylabel('Number of Unique Values')
    plt.title('Number of Unique Values in Categorical Features')
    plt.xticks(rotation=90)
    plt.show()
else:
    print("Error: Lengths of 'object_cols' and 'unique_values' do not match.")
```

```
import seaborn as sns
import matplotlib.pyplot as plt

object_cols = [col for col in dataset.columns if dataset[col].dtype == 'object']

unique_values = [dataset[col].nunique() for col in object_cols]

color_dict = {
    'MSZoning': 'blue',
    'LotConfig': 'green',
    'Exterior1st': 'orange',
    'BldgType': 'red'
}

colors = [color_dict.get(col, 'grey') for col in object_cols]

plt.figure(figsize=(12, 6))
sns.barplot(x=object_cols, y=unique_values, palette=colors)
plt.xlabel('Categorical Features')
plt.ylabel('Number of Unique Values')
plt.title('Number of Unique Values in Categorical Features')
plt.xticks(rotation=90)
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 36))
plt.suptitle('Categorical Features: Distribution', fontsize=20)
index = 1

for col in object_cols:
    y = dataset[col].value_counts()

    num_colors = len(y)
    colors = sns.color_palette("husl", num_colors)

    plt.subplot(11, 4, index)
    sns.barplot(x=list(y.index), y=y, palette=colors)

    plt.title(col)
    plt.xticks(rotation=90)
    plt.xlabel('Categories')
    plt.ylabel('Count')

    index += 1

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error

data = pd.read_excel("HousePricePrediction.xlsx")

X = data[['feature1', 'feature2', 'feature3']]  # Replace with your actual feature columns
Y = data['target']  # Replace with your actual target column

X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, test_size=0.2, random_state=42)

model_SVR = svm.SVR()
model_SVR.fit(X_train, Y_train)

Y_pred = model_SVR.predict(X_valid)

print(mean_absolute_percentage_error(Y_valid, Y_pred))
```
